### Perceptron 

```python
import torch
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

X, Y = torch.rand(100, 64).to(device), torch.rand(100,  1).to(device)

#Note: Ensure all are leaf tensors: https://stackoverflow.com/questions/65301875/how-to-understand-creating-leaf-tensors-in-pytorch
w1 = torch.rand(64, 32, requires_grad=True, device=device)
b1 = torch.rand(1, requires_grad=True, device=device)
w2 = torch.rand(32, 1, requires_grad=True, device=device)
b2 = torch.rand(1, requires_grad=True, device=device)

lr=1.e-3

optim = torch.optim.Adam([w1,b1,w2,b2], lr=lr)

for iter in range(1000):
    optim.zero_grad()
    h = torch.matmul(X, w1) + b1
    y = torch.matmul(h, w2) + b2
    loss = torch.nn.functional.binary_cross_entropy_with_logits(y.cpu(), Y.cpu())
    print(f'iter: {iter}//loss:{loss.item():0.3f}')
    loss.backward()
    optim.step()
```
